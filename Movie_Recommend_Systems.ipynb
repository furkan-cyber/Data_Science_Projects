```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies_df=pd.read_csv("movies.csv")
```

```python
movies_df.head()
```

```python
movies_df.shape
```

```python
selected_features=["genres","keywords","tagline","cast","director"]
print(selected_features)
```

```python
for feature in selected_features:
    movies_df[feature]=movies_df[feature].fillna("")
```

```python
combined_features=movies_df["genres"]+" "+movies_df["genres"]+" "+movies_df["keywords"]+" "+movies_df["tagline"]+" "+movies_df["cast"]+" "+movies_df["director"]
```

```python
print(combined_features)
```

```python
vectorizer=TfidfVectorizer()
```

```python
feature_vector=vectorizer.fit_transform(combined_features)
```

```python
print(feature_vector)
```

```python
similarity=cosine_similarity(feature_vector)
```

```python
print(similarity)
```

```python
print(similarity.shape)
```

```python
movie_name=input("Please enter movie name: ")
```

```python
list_of_all_titles=movies_df["title"].tolist()
print(list_of_all_titles)
```

```python
find_close_match=difflib.get_close_matches(movie_name,list_of_all_titles)
print(find_close_match)
```

```python
close_match=find_close_match[0]
print(close_match)
```

```python
index_of_the_movie=movies_df[movies_df.title==close_match]["index"].values[0]
index_of_the_movie
```

```python
similarity_score=list(enumerate(similarity[index_of_the_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
sorted_similar_movies=sorted(similarity_score,key=lambda x:x[1],reverse=True)
print(sorted_similar_movies)
```

```python
print('Movies suggested for you : \n')

i = 1

for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movies_df[movies_df.index==index]['title'].values[0]
  if (i<30):
    print(i, '.',title_from_index)
    i+=1
```

```python
movie_name = input(' Enter your favourite movie name : ')

list_of_all_titles = movies_df['title'].tolist()

find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)

close_match = find_close_match[0]

index_of_the_movie = movies_df[movies_df.title == close_match]['index'].values[0]

similarity_score = list(enumerate(similarity[index_of_the_movie]))

sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True) 

print('Movies suggested for you : \n')

i = 1

for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movies_df[movies_df.index==index]['title'].values[0]
  if (i<30):
    print(i, '.',title_from_index)
    i+=1
```
