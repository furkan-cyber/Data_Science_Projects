# Data Preprocessing

```python
df.info()
```

```python
df.head()
```

```python
df["reading score"].mean()
```

```python
df["writing score"].mean()
```

```python
df["math score"].mean()
```

```python
df["gender"].unique()
```

```python
df["race/ethnicity"].unique()
```

```python
df["parental level of education"].unique()
```

```python
df["lunch"].unique()
```

```python
df["test preparation course"].unique()
```

```python
# Radical variable not in data
```

```python
df["gender"]=df["gender"].map({"female":0,"male":1})
```

```python
df["race/ethnicity"]=df["race/ethnicity"].map({"group A":0,"group B":1,"group C":2,"group D":3,"group E":4})
```

```python
df["parental level of education"]=df["parental level of education"].map({"high school":0,
                                       "some high school":1,
                                       "some college":2,
                                       "associate's degree":3,
                                       "bachelor's degree":4,
                                       "master's degree":5})
```

```python
df["lunch"]=df["lunch"].map({"standard":0,"free/reduced":1})
```

```python
df["test preparation course"]=df["test preparation course"].map({"none":0,"completed":1})
```

```python
#ordinal variables
from pandas.api.types import CategoricalDtype
```

```python
df["parental level of education"]=df["parental level of education"].astype(CategoricalDtype(ordered=True,categories=[0,1,2,3,4,5]))
```

```python
df["test preparation course"]=df["test preparation course"].astype(CategoricalDtype(ordered=True,categories=[0,1]))
```

# Machine Learning

```python
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier
from xgboost import XGBClassifier
from sklearn.ensemble import GradientBoostingClassifier,RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV,train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.multioutput import MultiOutputClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
import numpy as np
```

```python
X=df.drop("gender",axis=1)
y=df["gender"]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
knn_=KNeighborsClassifier().fit(X_train,y_train)
rfc_=RandomForestClassifier().fit(X_train,y_train)
dct_=DecisionTreeClassifier().fit(X_train,y_train)
lgbm_=LGBMClassifier().fit(X_train,y_train)
lr_=LogisticRegression().fit(X_train,y_train)
```
