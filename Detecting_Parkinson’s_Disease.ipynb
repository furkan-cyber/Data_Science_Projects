```python
parkinson_df.drop(["name"],axis=1,inplace=True)
label=parkinson_df.columns
for i in label:
    if i!="status":
      q1=parkinson_df[i].quantile(0.25)
      q3=parkinson_df[i].quantile(0.75)
      Ä±qr=q3-q1
      lower=q1-(1.5*Ä±qr)
      upper=q3+(1.5*Ä±qr)
      lower_radical=(parkinson_df[i]<(lower))
      upper_radical=(parkinson_df[i]>(upper))
      parkinson_df[i][lower_radical]=lower
      parkinson_df[i][upper_radical]=upper
print("Data Cleaned")
```

```python
import numpy as np
def minmax(arr):
    l = np.min(arr)
    h = np.max(arr)
    return (arr -l) / (h-l) 
```

```python
df_2=parkinson_df.drop(["status"],axis=1)
minmax(df_2)
```

```python
df_2["status"]=parkinson_df["status"]
```
